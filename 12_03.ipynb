```python
class MinPriorityQueueHeap:
  def __init__ (self, lst = []):
    self.lst = lst
    for i in range (len (lst) - 1, -1, -1):
      self.bubbledownrec (i)

  def isEmpty (self):
    return len (self.lst) == 0

  def minimum (self):
    return self.lst[0]

  def bubbleup (self, idx):
    while idx != 0:
      paridx = (idx - 1) // 2
      if self.lst[idx] < self.lst[paridx]:
        self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
        idx = paridx
      else:
        return

  def insert (self, x):
    self.lst.append (x)
    self.bubbleup (len (self.lst) - 1)

  def bubbledownrec (self, idx):
    if 2 * idx + 1 >= len (self.lst):
      return
    smchld = 2 * idx + 1
    rchld = 2 * idx + 2
    if rchld < len (self.lst) and self.lst[rchld] < self.lst[smchld]:
      smchld = rchld

    if self.lst[smchld] < self.lst[idx]:
      self.lst[smchld], self.lst[idx] = self.lst[idx], self.lst[smchld]
      self.bubbledownrec (smchld)

  def extractmin (self):
    mn = self.lst[0]
    self.lst[0] = self.lst[-1]
    self.lst.pop ()
    self.bubbledownrec (0)
    return mn
```

```python
lst = [5, 10, 15, 123, 20, 70, 99, 50, 1, 27, 13, 99, 2]
```

```python
# We want to construct a min-heap of the elements in the lst

# Option A: we can sort the lst (ascending for min-heap,
# descending for max-heap)

mnhp1 = sorted (lst)
print (mnhp1)

# Option B: start with an empty min-heap, and insert all elements
# of lst one by one

mnhp2 = MinPriorityQueueHeap ()
for elem in lst:
  mnhp2.insert (elem)
print (mnhp2.lst)

# Option C: We consider the array as representing a binary tree
# and we perform bubble-down on every node, from the bottom level
# to the top

mnhp3 = MinPriorityQueueHeap ()
mnhp3.lst = lst
for i in range (len (lst) - 1, -1, -1):
  mnhp3.bubbledownrec (i)
print (mnhp3.lst)

# In practice, Option C is the fastest

# for this course, we will assume build-max-heap and build-min-heap
# refer to Option C: bubble-down every node, from bottom to top
```

```python
class MaxPriorityQueueHeap:
  def __init__ (self, lst = []):
    self.lst = lst
    for i in range (len (lst) - 1, -1, -1):
      self.bubbledownrec (i)

  def isEmpty (self):
    return self.lst == []
    return len (self.lst) == 0

  def maximum (self):
    # max element is always the root
    return self.lst[0]

  # assume all elements of the heap satisfy the
  # order property, except possibly the element
  # at index idx, which may violate the order
  # with its parent
  def bubbleup (self, idx):
    while idx != 0:
      paridx = (idx - 1) // 2
      if self.lst[idx] > self.lst[paridx]:
        self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
        idx = paridx
      else:
        return

  def bubbleuprec (self, idx):
    if idx == 0:
      return
    paridx = (idx - 1) // 2
    if self.lst[idx] > self.lst[paridx]:
      self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
      self.bubbleuprec (paridx)

  def insert (self, x):
    self.lst.append (x)
    self.bubbleup (len (self.lst) - 1)

  def bubbledown (self, idx):
    while 2 * idx + 1 < len (self.lst):  # while left child exists
      bgchld = 2 * idx + 1    # initialize bigger child to left
      if 2 * idx + 2 < len (self.lst):  # if right child exists
        if self.lst[2 * idx + 2] > self.lst[bgchld]:  # and is bigger than left
          bgchld = 2 * idx + 2

      if self.lst[bgchld] > self.lst[idx]:
        self.lst[bgchld], self.lst[idx] = self.lst[idx], self.lst[bgchld]
        idx = bgchld
      else:
        return

  def bubbledownrec (self, idx):
    if 2 * idx + 1 >= len (self.lst):
      return
    bgchld = 2 * idx + 1
    if 2 * idx + 2 < len (self.lst):  # if right child exists
      if self.lst[2 * idx + 2] > self.lst[bgchld]:  # and is bigger than left
        bgchld = 2 * idx + 2
    if self.lst[bgchld] > self.lst[idx]:
      self.lst[bgchld], self.lst[idx] = self.lst[idx], self.lst[bgchld]
      self.bubbledown (bgchld)

  def extractmax (self):
    mx = self.lst[0]              # save the root (maximum)
    self.lst[0] = self.lst[-1]    # move last node to root
    self.lst.pop ()
    self.bubbledown (0)
    return mx
```
