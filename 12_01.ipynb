```python
class MaxPriorityQueueHeap:
  def __init__ (self):
    self.lst = []

  def isEmpty (self):
    return self.lst == []
    return len (self.lst) == 0

  def maximum (self):
    # max element is always the root
    return self.lst[0]

  # assume all elements of the heap satisfy the
  # order property, except possibly the element
  # at index idx, which may violate the order
  # with its parent
  def bubbleup (self, idx):
    while idx != 0:
      paridx = (idx - 1) // 2
      if self.lst[idx] > self.lst[paridx]:
        self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
        idx = paridx
      else:
        return

  def bubbleuprec (self, idx):
    if idx == 0:
      return
    paridx = (idx - 1) // 2
    if self.lst[idx] > self.lst[paridx]:
      self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
      self.bubbleuprec (paridx)

  def insert (self, x):
    self.lst.append (x)
    self.bubbleup (len (self.lst) - 1)

  def bubbledown (self, idx):
    while 2 * idx + 1 < len (self.lst):  # while left child exists
      bgchld = 2 * idx + 1    # initialize bigger child to left
      if 2 * idx + 2 < len (self.lst):  # if right child exists
        if self.lst[2 * idx + 2] > self.lst[bgchld]:  # and is bigger than left
          bgchld = 2 * idx + 2

      if self.lst[bgchld] > self.lst[idx]:
        self.lst[bgchld], self.lst[idx] = self.lst[idx], self.lst[bgchld]
        idx = bgchld
      else:
        return

  def bubbledownrec (self, idx):
    if 2 * idx + 1 >= len (self.lst):
      return
    bgchld = 2 * idx + 1
    if 2 * idx + 2 < len (self.lst):  # if right child exists
      if self.lst[2 * idx + 2] > self.lst[bgchld]:  # and is bigger than left
        bgchld = 2 * idx + 2
    if self.lst[bgchld] > self.lst[idx]:
      self.lst[bgchld], self.lst[idx] = self.lst[idx], self.lst[bgchld]
      self.bubbledown (bgchld)

  def extractmax (self):
    mx = self.lst[0]              # save the root (maximum)
    self.lst[0] = self.lst[-1]    # move last node to root
    self.lst.pop ()
    self.bubbledown (0)
    return mx
```

```python
maxpq = MaxPriorityQueueHeap ()
maxpq.insert (35)
maxpq.insert (87)
maxpq.insert (42)
print (maxpq.maximum ())    # 87
print (maxpq.extractmax ()) # 87
print (maxpq.maximum ())    # 42
maxpq.insert (65)
print (maxpq.maximum ())    # 65
maxpq.extractmax ()
print (maxpq.maximum ())       # 42
print (maxpq.extractmax ()) # 42
maxpq.insert (21)
print (maxpq.maximum ())    # 35
maxpq.extractmax ()
print (maxpq.isEmpty ())    # False
print (maxpq.extractmax ()) # 21
print (maxpq.isEmpty ())    # True
```

```python
# In a min-heap, the structural property is the same
# (all levels full, except possibly the last level, which is still
# filled from the left)

# Order property is the opposite: for every node, the parent is smaller
# or equal
```

```python
class MinPriorityQueueHeap:
  def __init__ (self):
    self.lst = []

  def isEmpty (self):
    return len (self.lst) == 0

  def minimum (self):
    return self.lst[0]

  def bubbleup (self, idx):
    while idx != 0:
      paridx = (idx - 1) // 2
      if self.lst[idx] < self.lst[paridx]:
        self.lst[idx], self.lst[paridx] = self.lst[paridx], self.lst[idx]
        idx = paridx
      else:
        return

  def insert (self, x):
    self.lst.append (x)
    self.bubbleup (len (self.lst) - 1)

  def bubbledownrec (self, idx):
    if 2 * idx + 1 >= len (self.lst):
      return
    smchld = 2 * idx + 1
    rchld = 2 * idx + 2
    if rchld < len (self.lst) and self.lst[rchld] < self.lst[smchld]:
      smchld = rchld

    if self.lst[smchld] < self.lst[idx]:
      self.lst[smchld], self.lst[idx] = self.lst[idx], self.lst[smchld]
      self.bubbledownrec (smchld)

  def extractmin (self):
    mn = self.lst[0]
    self.lst[0] = self.lst[-1]
    self.lst.pop ()
    self.bubbledownrec (0)
    return mn
```

```python
mnhp = MinPriorityQueueHeap ()
mnhp.insert (21)
mnhp.insert (31)
mnhp.insert (50)
mnhp.insert (77)
mnhp.insert (3)
mnhp.insert (7)
mnhp.insert (34)
mnhp.insert (17)

print (mnhp.lst)

mnhp.extractmin ()
mnhp.extractmin ()

print (mnhp.lst)
```

```python
minpq = MinPriorityQueueHeap ()
minpq.insert (35)
minpq.insert (87)
minpq.insert (42)
print (minpq.minimum ())    # 35
print (minpq.extractmin ()) # 35
print (minpq.minimum ())    # 42
minpq.insert (65)
print (minpq.minimum ())    # 42
minpq.extractmin ()
print (minpq.minimum ())       # 65
print (minpq.extractmin ()) # 65
minpq.insert (21)
print (minpq.minimum ())    # 21
minpq.extractmin ()
print (minpq.isEmpty ())    # False
print (minpq.extractmin ()) # 87
print (minpq.isEmpty ())    # True
```
