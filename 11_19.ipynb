```python
# A graph describes a scenario where we have a collection
# of data elements, but some elements have some kind of
# connection with other elements

# Examples: friends/followers in a social networking site,
# transportation network
```

```python
# Each data element is referred to as a vertex (plural: vertices)
# and a connection between two elements is referred to as an edge
```

```python
# Graph Classifications:

# Undirected vs Directed: For an undirected graph, an edge from u to
# v is the same as an edge from v to u (every edge goes both ways).
# For a directed graph, an edge from u to v is NOT the same as an edge
# from v to u. It's even possible to have one without the other.

# Unweighted vs Weighted: In a weighted graph, every edge has some
# value known as a weight. In an unweighted graph, we treat all edges
# as having the same weight.
```

```python
# For a directed graph with N vertices, what is the maximum possible
# number of edges? N(N - 1)

# For a undirected graph with N vertices, what is the maximum possible
# number of edges? N(N - 1)/2


# Sparse vs Dense: A graph is sparse if the number of edges is less
# than half of the maximum possible. A graph is dense if the number
# of edges is more than half of the maximum possible.

# A complete graph is a graph with the maximum possible number of
# edges
```

```python
# A connected graph is a graph where there is a path between
# any two vertices

# For a directed graph, we say it is strongly connected if there
# is a path from any vertex to any vertex (following the directions)
# We say it is weakly connected if the graph is connected after
# ignoring the directions.
```

```python
# In an undirected graph, the degree of a vertex is the number of
# edges connected to that vertex
# Sum of degrees = 2 x Number of Edges

# In a directed graph, there are two notions of degree:
# outdegree and indegree

# The outdegree of a vertex is the number of outgoing edges from that
# vertex.
# The indegree of a vertex is the number of incoming edges towards
# that vertex.
# Sum of all outdegrees = Sum of all indegrees = Number of Edges
```
