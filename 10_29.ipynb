```python
# How to write recursive code

# My suggestion: just make sure to follow THREE essential rules
```

```python
# RULE NUMBER ONE: There must be at least one base case

# We can implement very simple cases as base cases
# Cases that are simple enough that we can manually deal with
# them without recursion
# e.g., if we have to do something with a list, we can use the
# empty list as a base case, or a list with only one element,
# empty strings for strings, small numbers like 0 or 1 for numbers, etc
```

```python
# RULE NUMBER TWO: Every recursive call on a valid argument MUST
# eventually reach a base case (even if the base case is on an
# invalid argument)
# We only need to consider what the arguments are to the recursive
# calls to verify rule 2

# Assume that x is a nonnegative integer
def fun1 (x):
  if x == 0:
    return 7
  return fun1 (x - 1) + 3

# This satisfies rule 1 and rule 2
```

```python
# Assume that x is a nonnegative integer
def fun2 (x):
  if x == 0:
    return 7
  return fun2 (x + 1) + 3

# This does not satisfy rule 2
```

```python
# Assume that x is a nonnegative integer
def fun3 (x):
  if x == 0:
    return 7
  if x == 3000000:
    return 91
  return fun3 (x + 1) + 3

# This does not satisfy rule 2, e.g., x = 4000000 will not
# reach any base case
```

```python
# Assume that x is a nonnegative integer
def fun4 (x):
  if x == 0:
    return 7
  return fun4 (x - 2) + 3

# This does NOT satisfy rule 2, because odd numbers will not
# reach any base cases
```

```python
# Assume that x is a nonnegative integer
def fun5 (x):
  if x <= 0:
    return 7
  return fun5 (x - 2) + 3

# This satisfies rule 2
```

```python
# Assume that x is a nonnegative integer
def fun6 (x):
  if x == 0:
    return 7
  if x == 1:
    return 4
  return fun6 (x - 2) + 3

# This satisfies rule 2
```

```python
# Assume that x is a nonnegative integer
def fun7 (x):
  if x == 0:
    return 7
  if x == 1:
    return 4
  if x == 2:
    return 9
  if x == 3:
    return 8
  return fun7 (x - 5) + 3

# This does not satisfy rule 2
```

```python
# Assume that x is a nonnegative integer
def fun8 (x):
  if x == 0:
    return 7
  if x == 1:
    return 4
  if x == 2:
    return 9
  if x == 3:
    return 8
  if x == -71:
    return 5
  return fun8 (x - 5) + 3

# This satisfies rule 2
# In general, if a recursive call is called on x - T, then
# we need at least T base cases
```

```python
# RULE NUMBER THREE: Imagine that every recursive call will connect
# to an AI supercomputer which is 100% guaranteed to perform the
# correct task and return the correct answer
# You can then exploit this to achieve your required task
```

```python
# Write a recursive function to add all elements of a list of integers

def getsum1 (lst):
  if len (lst) == 0:
    return 0
  if len (lst) == 1:
    return lst[0]
  aians = getsum1 (lst)
  return aians

# This satisfies rule 3 and rule 1, but NOT rule 2

print (getsum1 ([8, 3, 1, 4]))    #
```

```python
# If you try to make the AI supercomputer do ALL of the work,
# then this will usually fail rule 2

# But if I can do even a bit of work to make the argument even slightly
# closer to a base case, then it should satisfy rule 2
```

```python
# Write a recursive function to add all elements of a list of integers
def getsum2 (lst):
  if len (lst) == 0:
    return 0
  ailst = lst[1:]     # ailst has everything in lst except 1st element
  aians = getsum2 (ailst)   # aians is the sum of all elements of ailst
  return lst[0] + aians

# this satisfies rule 1 and rule 2 and rule 3
print (getsum2 ([8, 3, 1, 4]))    # 16
```

```python
# Write a recursive function to multiply all elements of a
# non-empty list of integers

def getmul (lst):
  if len (lst) == 1:
    return lst[0]
  ailst = lst[1:]
  aians = getmul (ailst)
  return lst[0] * aians

print (getmul ([8, 1, 3, 4]))   # 96
```

```python
# Compute n! = 1 x 2 x 3 x 4 x ... x n using recursion
# Note: 0! = 1
# Assume n is a nonnegative integer

def factorial (n):
  if n == 0:
    return 1
  aians = factorial (n - 1)
  return aians * n

print (factorial (5))   # 120
```
